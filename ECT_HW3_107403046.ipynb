## 2.

### 2.1 載入csv

```python
import pandas as pd
df = pd.read_csv('Desktop/ECT_HW3/train.csv')
df
```

### 2.2 檢查空值個數

```python
df.isnull().sum()
```

### 2.3

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Cabin'] = df['Cabin'].fillna(0)
df.Cabin[df['Cabin'] != 0] = 1
df
```

### 2.4
將男性設為0女性設為1

將S設為0 C設為1 Q設為2

```python
df.Sex[df['Sex'] == 'male'] = 0
df.Sex[df['Sex'] == 'female'] = 1

df.Embarked[df['Embarked'] == 'S'] = 0
df.Embarked[df['Embarked'] == 'C'] = 1
df.Embarked[df['Embarked'] == 'Q'] = 2
df['Embarked'] = df['Embarked'].fillna(3)

df
```

### 2.5

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

x = df[['Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked']]
y = df[['Survived']]

X_train,X_test,y_train,y_test=train_test_split(x,y, test_size=0.5, random_state=12,stratify=y)
model = DecisionTreeClassifier(criterion='gini', max_depth=4)
train = model.fit(X_train, y_train)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
tree.plot_tree(train, fontsize=12)
plt.show()
```

### 2.6

```python
score50 = model.score(X_test, y_test)
score50
```

### 2.7

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.6, random_state=12,stratify=y)
train60 = model.fit(X_train, y_train)
score60 = model.score(X_test, y_test)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.7, random_state=12,stratify=y)
train70 = model.fit(X_train, y_train)
score70 = model.score(X_test, y_test)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.8, random_state=12,stratify=y)
train80 = model.fit(X_train, y_train)
score80 = model.score(X_test, y_test)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.9, random_state=12,stratify=y)
train90 = model.fit(X_train, y_train)
score90 = model.score(X_test, y_test)

data = {
    'split_proportion':['50/50','60/40','70/30','80/20','90/10'],
    'score':[score50,score60,score70,score80,score90]
}
scoredf = pd.DataFrame(data)
scoredf.head()
```

### 2.8

```python
plt.plot(scoredf["split_proportion"], scoredf["score"],c = "b")  

plt.xlabel("train/test proportion", fontweight = "bold")
plt.title("train/test split proportion & result", fontsize = 15, fontweight = "bold", y = 1.1)
plt.xticks(rotation=90)
plt.show()
```
